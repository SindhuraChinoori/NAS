```python
import pandas as pd #importing libraries
pupil_marks = pd.read_csv("E:/gramener/training/nas-pupil-marks.csv") # reading pupil data from csv file
pupil_marks.head() #printing head
```

```python
import numpy as np
pupil_marks['Total %'] = pupil_marks[['Maths %','Reading %','Science %','Social %']].mean(axis=1)

pupil_marks['Distance'].replace([0,2,3,4,1],[np.nan,'1-3 km','3-5 km','More than 5 km','Up to 1 km'],inplace=True)
# pupil_marks[['Maths %','Reading %','Science %','Social %']].sum(axis=1)
pupil_marks.head()
```

```python
f = pupil_marks.groupby(['Distance']).agg({'Maths %': 'mean','Reading %': 'mean','Science %': 'mean','Social %': 'mean','Total %': 'mean'})
# f.max() - f.min()
# f.max()
# pd.DataFrame([f.max(), f.min()], columns=["max", "min"])
f
df = pd.DataFrame(f)
```

```python

# df['Maths %']  = df['Maths %'] - df['Maths %'].min()
cols_of_interest = list(df.columns.values)
# df.rename()
df2 = df[cols_of_interest] - df[cols_of_interest].min()
df2
```

```python
# import matplotlib
# matplotlib.use('Agg')
import matplotlib.pyplot as plt
import seaborn as sns

# ax = sns.barplot(x="Total %", data=df)
# plt.figure(figsize=(8,4))
for col in df.columns.values:
    fig, ax =plt.subplots(ncols=2,figsize=(15,4))
    sns.barplot(x=col, y = df2.index, data=df2,ax=ax[0])

    # ax.set(xlim=(0.5,4.5))
    # ax[0].set_ylabel('distance')
    ax[0].set_xlabel('Impact of distance on '+col)
    #plt.show()
    sns.barplot(y=col, x = df.index, data=df, ax=ax[1])
    
    plt.xticks(rotation=45)

    plt.show()
```
